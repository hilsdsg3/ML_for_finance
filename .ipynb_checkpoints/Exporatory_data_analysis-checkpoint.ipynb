# Theory :

### Stock Price Dynamics :

Modeling stock price dynamics is an important aspect of mathematical finance.  Simulating stock price movements allows an algorithmic trader to test different strategies. 
An important aspect of every financial model it's limitations and assumptions.

### Geometric Brownian Motion :
Simulations of stocks and other security instruments are modeled using stochastic differential equations (SDEs).  
Because of the randomness associated with stock price movements, the models cannot be developed using ordinary differential equations (ODEs).

A typical model used for price dynamics is the following stochastic differential equation:

$dS = \mu S\ dt +\sigma S\ dW_t $

where $S$  is the stock price, $\mu$ is the drift coefficient, $\sigma$ is the diffusion coefficient, and $W_t$ is the [Brownian Motion](https://en.wikipedia.org/wiki/Wiener_process).  

In modeling a stock price, the drift coefficient represents the mean of returns over some time period, and the diffusion coefficient represents the standard deviation of those same returns.  

The Brownian Motion $W_t$ is the randomized portion of the above equation.  Each Brownian motion $W_i$ is computed by multiplying a standard random variable $z_i$ from a normal Gaussian distribution $N(0,1)$ with mean $0$ and standard deviation $1$ by the square root of the time increment $\sqrt{\Delta t_i}$. $$ $$
$ W_i = z_i \sqrt{\Delta t_i}$

The cumulative sum of the Brownian increments is the discrete Brownian path.

$W_{n}(t) = \sum_{i=1}^{n} W_i(t)$

For the SDE above with an initial condition for the stock price of $S(0) = S_{0}$, the closed form solution of [Geometric Brownian Motion](https://en.wikipedia.org/wiki/Geometric_Brownian_motion) (GBM) is:  

$ S(t) = S_{0}e^{(\mu-\frac{1}{2}\sigma^2)t+\sigma W_t}$

### Brownian Motion in Python :

Before one can model the closed form solution of GBM, we need to model the Brownian Motion.  This is the stochastic portion of the equation.  To do this we'll need to generate the standard random variables from the normal distribution $N(0,1)$.  Next, we'll multiply the random variables by the square root of the time step.  The final step is to performa cumulative sum to generate the Brownian Motion.


# Setup , import , and functions :

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas_datareader import data as pdr
from plotly.offline import init_notebook_mode, iplot, plot
import chart_studio.plotly as py
import plotly.graph_objs as go
import pathlib
FILE_FOLDER = 'C:/Users/champ/Python_proj/ML_training_repo/Fundamentals/meta_data/'
OUTPUT_UNSTRUCTURED_FILE = 'unstructured_EDA.csv'
OUTPUT_STRUCTURED_FILE = 'structured_EDA.csv'

def Brownian(N, random_seed = False, random_seed_nr = None):
    
    if random_seed:
        r = np.random.RandomState(random_seed_nr)                        
        dt = 1./N                                    # time step
        b = r.normal(0., 1., int(N))*np.sqrt(dt)  # brownian increments
    else:
        dt = 1./N                                    # time step
        b = np.random.normal(0., 1., int(N))*np.sqrt(dt)  # brownian increments
    W = np.cumsum(b)                             # brownian path
    return W, b

def GBM(starting_value, mu, sigma, W, T, N):  
    '''
    # GBM Exact Solution
    # Parameters
    #
    # starting_value :     initial stock price
    # mu:     returns (drift coefficient)
    # sigma:  volatility (diffusion coefficient)
    # W:      brownian motion
    # T:      time period
    # N:      number of increments
    '''
    
    t = np.linspace(0.,1.,int(N+1))
    S = []
    S.append(starting_value)
    for i in range(1,int(N+1)):
        drift = (mu - 0.5 * sigma**2) * t[i]
        diffusion = sigma * W[i-1]
        S_temp = starting_value*np.exp(drift + diffusion)
        S.append(S_temp)
    return S, t

def daily_return(adj_close):
    returns = []
    for i in range(0, len(adj_close)-1):
        today = adj_close[i+1]
        yesterday = adj_close[i]
        daily_return = (today - yesterday)/yesterday
        returns.append(daily_return)
    return returns

# Euler Maruyama Approximation
def EM(starting_value, mu, sigma, b, T, N, M):
    dt = M * (1/N)  # EM step size
    L = N / M
    wi = [starting_value]
    for i in range(0,int(L)):
        Winc = np.sum(b[(M*(i-1)+M):(M*i + M)])
        w_i_new = wi[i]+mu*wi[i]*dt+sigma*wi[i]*Winc
        wi.append(w_i_new)
    return wi, dt

def plotly_time_series(df_labels,df_titles,df, font = 2):
    count = 0
    d = {} #Empty dictionary to add values into
    list_dict_titles = list(df_titles.keys())
    list_dict = list(df_labels.keys())
    print(list_dict)
    for i in list_dict:
        d[count] = go.Scatter(x=df.index,
                           y=df[df_labels[i]['column_name']],
                           line=go.scatter.Line(color=df_labels[i]['color'], width = font*3),
                               opacity=0.8,
                               name=df_labels[i]['name'],
                               text=df_labels[i]['text'])
        count+=1
    layout = go.Layout(height=800, width=1400, font=dict(size=20),
                       title=df_titles['titles']['chart_title'],
                        yaxis=dict(title=df_titles['titles']['y_axis_title'], showspikes=True,
                                   fixedrange = False),
                       xaxis=dict(title=df_titles['titles']['x_axis_title'], showspikes=True,
                               rangeslider=dict(visible=True),
                               rangeselector=dict(
                               buttons=list([
                            dict(count=1, label="1m", step="month", stepmode="backward"),
                            dict(count=6, label="6m", step="month", stepmode="backward"),
                            dict(count=1, label="YTD", step="year", stepmode="todate"),
                            dict(count=1, label="1y", step="year", stepmode="backward"),
                            dict(count=6, label="5y", step="year", stepmode="backward"),
                            dict(step="all")
                    ]))))
    tot_temp = ''
    for i in range(len(d)):
        temp = 'd[' + str(i) + ']'
        comma = ','
        if i < len(d):
            tot_temp = tot_temp + temp + comma
        else:
            tot_temp = tot_temp + temp
    fig = go.Figure(data=eval(tot_temp), layout=layout)
    iplot(fig)
```

# Structered stock data generation :

```python
# Download of stock data and generation of stock data

data_type = 'Adj Close'
stock = 'TSLA'
T = 1.

start_date = '2014-01-02'
end_date = '2019-11-14'
random_seed_nr = 100
rndm_seed = True


df = pdr.get_data_yahoo([stock], start=start_date, end=end_date)[data_type]
initial_value = df[stock][0]
N = len(df)

returns = daily_return(df[stock])
mu_st = np.mean(returns)*252.          # drift coefficient
sig_st = np.std(returns)*np.sqrt(252.)  # diffusion coefficient

# Generation of simulated stock data
W = Brownian(N, random_seed = rndm_seed, random_seed_nr = random_seed_nr)[0]
soln = GBM(initial_value, mu, sig, W, T, N)[0]    # Exact solution

soln = soln[:-1]
df['GBM'] = soln

returns = daily_return(df['GBM'])
mu_GBM_st = np.mean(returns)*252.           # drift coefficient
sig_GBM_st = np.std(returns)*np.sqrt(252.)  # diffusion coefficient

stck_nm = stock + ' mu=' + str(round(mu_st,4)) + ' sig=' +  str(round(sig_st,4))
GBM_nm = 'GBM' + ' mu=' + str(round(mu_GBM_st,4)) + ' sig=' +  str(round(sig_GBM_st,4))

df_labels = { stock:
      { "column_name": stock,
        "color": "gray",
        "name": stck_nm,
        "text": stock
      },'GBM': 
      { "column_name": "GBM",
        "color": "red",
        "name": GBM_nm,
        "text": "Brownian Motion"
      }}

plotly_titles = { 'titles':
      { 'chart_title': 'STRUCTERED timestamps of stock price vs simulated data',
        'y_axis_title': 'Prices',
        'x_axis_title': 'Date',
      }}

econ = plotly_time_series(df_labels,plotly_titles,df, font = .5)

df.to_csv(FILE_FOLDER + OUTPUT_STRUCTURED_FILE, sep=';')
```

# Unstructered stock data generation

```python
# Create unstructered time series

np.random.seed(random_seed_nr)

remove_n = int(len(df)*.6)
drop_indices = np.random.choice(df.index, remove_n, replace=False)
df_subset = df.drop(drop_indices)
df_subset.head(10)

initial_value = df_subset[stock][0]
N = len(df_subset)

returns = daily_return(df_subset[stock])
mu_un = np.mean(returns)*(len(df_subset)*252/len(df))         # drift coefficient
sig_un = np.std(returns)*np.sqrt((len(df_subset)*252/len(df)))  # diffusion coefficient

# # Generation of simulated stock data
# W = Brownian(N, random_seed = rndm_seed, random_seed_nr = random_seed_nr)[0]
# soln = GBM(initial_value, mu, sig, W, T, N)[0]    # Exact solution

# soln = soln[:-1]
# df_subset['GBM'] = soln

returns = daily_return(df_subset['GBM'])
mu_GBM_un = np.mean(returns)*252.          # drift coefficient
sig_GBM_un = np.std(returns)*np.sqrt(252.)  # diffusion coefficient

stck_nm = stock + ' mu=' + str(round(mu_un,4)) + ' sig=' +  str(round(sig_un,4))
GBM_nm = 'GBM' + ' mu=' + str(round(mu_GBM_un,4)) + ' sig=' +  str(round(sig_GBM_un,4))

df_labels = { stock:
      { "column_name": stock,
        "color": "gray",
        "name": stck_nm,
        "text": stock
      },'GBM': 
      { "column_name": "GBM",
        "color": "red",
        "name": GBM_nm,
        "text": "Brownian Motion"
      }}

plotly_titles = { 'titles':
      { 'chart_title': 'UN-STRUCTERED timestamps of stock price vs simulated data',
        'y_axis_title': 'Prices',
        'x_axis_title': 'Date',
      }}

econ = plotly_time_series(df_labels,plotly_titles,df_subset, font = .5)

df_subset.to_csv(FILE_FOLDER + OUTPUT_UNSTRUCTURED_FILE, sep=';')
```
